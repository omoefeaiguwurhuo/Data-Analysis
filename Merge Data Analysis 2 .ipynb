```python
# Select only numeric columns
numeric_data = Merge_Data.select_dtypes(include='number')
```

```python
numeric_data
```

```python
correlation=numeric_data.corr()
```

```python
correlation
```

```python
# Create a heatmap using seaborn
plt.figure(figsize=(8, 6))
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Plot')
plt.show()
```

```python
# Extracting column names excluding 'Weight'
features = numeric_data.columns[numeric_data.columns != 'weight']
 
# Plot scatter plots for each feature against 'Weight'
plt.figure(figsize=(12, 8))
for feature in features:
    plt.scatter(numeric_data[feature],numeric_data['weight'], label=feature, alpha=0.7)
 
plt.title('Scatter Plots of Weight Against Other Variables')
plt.xlabel('Variables')
plt.ylabel('Weight')
plt.legend()
plt.show()
```

```python
features
```

```python
# Scatter plot of 'Weight' against 'Age'
plt.figure(figsize=(15, 10))
sns.scatterplot(x='age', y='weight', data=numeric_data)
plt.title('Scatter Plot of Weight Against Age')
plt.xlabel('Age')
plt.ylabel('Weight')
plt.show()
```

```python
# Loop through columns and plot histograms
for column in numeric_data.columns:
    plt.figure(figsize=(8, 6))
    plt.hist(numeric_data[column], bins=25, edgecolor='black')
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```

```python
features = numeric_data.columns[numeric_data.columns != 'Weight']
 
# Loop through columns and plot scatter plots against 'Weight'
for feature in features:
    plt.figure(figsize=(8, 6))
    plt.scatter(numeric_data[feature], numeric_data['weight'])
    plt.title(f'Scatter Plot of {feature} Against weight')
    plt.xlabel(feature)
    plt.ylabel('weight')
    plt.show()
```

```python
# Select all columns except 'Weight' and 'Age'
features = numeric_data.columns.difference(['weight', 'age'])
 
# Loop through columns and create bar plots against 'Weight', color-coded by 'Age'
for feature in features:
    plt.figure(figsize=(10, 6))
    sns.barplot(x=feature, y='weight', hue='age', data=numeric_data)
    plt.title(f'Bar Plot of {feature} Against Weight (Color by Age)')
    plt.xlabel(feature)
    plt.ylabel('weight')
    plt.legend(title='age')
    plt.show()
```
